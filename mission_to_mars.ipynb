```python
page_url = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"

page_response = requests.get(page_url)

page_text = page_response.text

page_soup = bs(page_text, "lxml")
```

```python
hemisphere_names = []

for div_tag in page_soup.find_all('div', "description"):
    
    hemi_text = div_tag.find("h3").text
    print(hemi_text)
    hemisphere_names.append(hemi_text)

print()
print("hemisphere_names")
print(hemisphere_names)
```

```python
hemisphere_links = []

for link in page_soup.find_all('a', 'itemLink'):
    
    link_href = link['href']
    print(link_href)
    hemisphere_links.append(link_href)

print()
print("hemisphere_links")
print(hemisphere_links)
```

```python
website = "https://astrogeology.usgs.gov"
image_links = []

for link in hemisphere_links:

    hemi_page_url = website + link
    hemi_page_response = requests.get(hemi_page_url)
    hemi_page_text = hemi_page_response.text
    hemi_page_soup = bs(hemi_page_text, "lxml")

    for div_tag in hemi_page_soup.find_all('div', 'downloads'):
        
        image_tag = div_tag.find('a', text ='Original')
        image_link = image_tag['href']
        print(image_link)
        image_links.append(image_link)

print()
print("image_links")
print(image_links)
    
```
